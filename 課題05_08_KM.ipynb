# ****DQ3風バトルシミュレータ****

## **概要**

DQ3に出てくる敵や魔法,武器を用いたバトルシミュレータです。

**出現する敵**

lev = 1 : 'スライム','いっかくウサギ','ホイミスライム','さまようよろい','キングスライム','メタルスライム','はぐれメタル'

lev = 2 : 'さまようよろい','ベホマスライム','キングスライム','カンダタこぶん','トロール','はぐれメタル','メタルキング'

lev = 3 : 'ベホマスライム','カンダタ','キングスライム','トロール','ボストロール','はぐれメタル','メタルキング'

## **操作方法**

1.ユーザ名入力

2.敵の強さを入力

3.バトルコマンドを入力する

4.どちらかのHPが0になるまで続ける
## **制作者のコメント**

ホイミスライムとベホイミスライムの行動分岐の実装にはとても苦労しました。また、状態異常や、バフ/デバフについては今回実装していませんが、また時間があるときに実装してみたいと思っています。

## **作成時間**

作成時間：**約16時間**

```python
%reset -f
import random as r_mod
import time

en1 = ['スライム','いっかくウサギ','ホイミスライム','さまようよろい','キングスライム','メタルスライム','はぐれメタル']
en2 = ['さまようよろい','ベホマスライム','キングスライム','カンダタこぶん','トロール','はぐれメタル','メタルキング']
en3 = ['ベホマスライム','カンダタ','キングスライム','トロール','ボストロール','はぐれメタル','メタルキング']
enm = ["メタルスライム","はぐれメタル","メタルキング"]
enhoim = ["ホイミスライム","ベホマスライム"]
per = [30,25,15,12,8,6,4]
enh = [100,120,120,130,140]
ena = [1,3,3,5,8,2,4]
b = ['先攻','後攻']
command = ['攻撃','呪文']
itper = [40,25,15,12,5,3]
item1 = ["ひのきの棒","はやぶさのけん","こんぼう","どうのつるぎ","せいなるナイフ","くさりがま"]
item2 = ["はやぶさのけん","どうのつるぎ","ブーメラン","やいばのブーメラン","てつのやり","はがねのつるぎ"]
item3 = ["やいばのブーメラン","はがねのつるぎ","てつのオノ","ほのおのブーメラン","ドラゴンテイル","いなずまのけん"]
items = {
    "ひのきの棒":2,"はやぶさのけん":5,"こんぼう":7,"どうのつるぎ":12,
    "せいなるナイフ":14,"くさりがま":16,"ブーメラン":19, "やいばのブーメラン":24,
    "てつのやり":26,"はがねのつるぎ":33,"てつのオノ":37,"ほのおのブーメラン":42,
    "ドラゴンテイル":52,"いなずまのけん":82
}
maho =[['メラ','イオ','バギ','ホイミ'],['メラミ','イオラ','ヒャド','バギマ','ベホイミ'],['メラゾーマ','イオナズン','マヒャド','バギクロス','ベホイム']]
#mera = ['メラ','メラミ','メラゾーマ']
#io = ['イオ','イオラ','イオナズン']
hoim = ['ホイミ','ベホイミ']
#hyad = ['ヒャド','マヒャド']
pl_sm = {
    'メラ':8, 'イオ':14, 'バギ':11,'ホイミ':30,
    'メラミ':56,'イオラ':48,'ヒャド':25,'バギマ':32,'ベホイミ':75,
    'メラゾーマ':160,'イオナズン':120,'ヒャダイン':68,'バギクロス':90,'ベホイム':164
}
pl_bg = {
    'メラ':13, 'イオ':21, 'バギ':26,'ホイミ':40,
    'メラミ':70,'イオラ':64,'ヒャド':35,'バギマ':55,'ベホイミ':95,
    'メラゾーマ':190,'イオナズン':160,'マヒャド':124,'バギクロス':120,'ベホイム':184
}

plname = input("プレイヤーネームを入力してください。")

x = 1
while x == 1:
  lev = int(input("敵の強さを選択してください。強さは1～3の整数値にしてください。"))
  #print(type(lev) is int)
  if type(lev) is int and lev >= 1 and lev <=3:
    x = 0
  else:
    print("無効な入力値です")

bn = r_mod.choice(b)
s = r_mod.choice(enh)
#print(s)
eh = s * 2 * lev /5
#print(f"eh:{eh}")

if lev == 1:
  ene = en1
  item = item1
elif lev == 2:
  ene = en2
  item = item2
elif lev ==3:
  ene = en3
  item = item3
#print(ene)

r = r_mod.choices(ene,weights=per,k = 3)
t = r_mod.choices(item,weights=itper,k = 3)

print("戦う敵を選んでください。")
for i in range(3):
  print(f'{i+1}.{r[i]}')

x = 1
while x == 1:
  stn = int(input("入力"))
  #print(type(stn) is int)
  if type(stn) is int and stn >= 1 and stn <= 3:
    x = 0
  else:
    print("無効な入力値です")

print("使う武器を選んでください。")
for i in range(3):
  print(f'{i+1}.{t[i]}')

x = 1
while x == 1:
  at = int(input("入力"))
  #print(type(stn) is int)
  if type(at) is int and at >= 1 and at <= 3:
    x = 0
  else:
    print("無効な入力値です")

enemy = r[stn-1]
a = ene.index(enemy)
#print(f"enemy:{enemy},a:{a}")
enatk = int(ena[a]) * lev
#print(f"enatk:{enatk}")
hp =r_mod.randint(100,240) + 12 * lev

print(f"あなたは{bn}です。")

#ステータス表示
time.sleep(0.1)
print(f"{enemy}----------------")
time.sleep(0.2)
print(f"HP = {eh}")
time.sleep(0.2)
print(f"ATK = {enatk}")
time.sleep(0.2)
print(f"{plname}----------------")
time.sleep(0.2)
print(f"HP = {hp}")
time.sleep(0.2)
print(f"ATK = {4-lev}")
time.sleep(0.2)
print(f"arm = {t[at-1]}")
time.sleep(0.2)
print(f"--------------------")
time.sleep(0.2)

for i in range(3):
  print(3 - i ,end = '')
  time.sleep(1)
print("")
print("S T A R T")

mt = 0
hoim = 2

for i in range(3):
  if enemy == enm[i]:
    mt = 1
  elif mt != 1:
    mt = 0
  if i != 2 and enemy == enhoim[i]:
    hoim = i
  elif hoim != 0:
    hoim = 2

if bn == "後攻":
  hp = hp - enatk
  print(f"{enemy}の攻撃！")
  print(f"{enatk}のダメージを受けた")

x = 1
while x == 1:
  y = 1
  print("あなたの番です。コマンド？")
  for i in range(2):
      print(f"{i+1}.{command[i]}")
  while y == 1:
    com = int(input(""))
    if type(com) is int and com >= 1 and com <=2:
      if com == 2:
        e = len(maho[lev-1])
        for i in range(e):
          print(f"{i+1}.{maho[lev-1][i]}")
        z = 1
        while z == 1:
          com2 = int(input(""))
          if type(com2) is int and com2 >= 1 and com2 <=int(e):
            z = 0
          else:
            print("無効な入力値です")
        dam1 = pl_sm[maho[lev-1][com2-1]]
        dam2 = pl_bg[maho[lev-1][com2-1]]
        dam = r_mod.randint(dam1, dam2)
      else:
        dam = items[t[at-1]]*(4-lev)
      y = 0
    else:
      print("無効な入力値です")

    if mt == 1:
      dam = int(dam / 4)

    if com == 1:
      eh = eh - dam
      print(f"{plname}の攻撃！")
      print(f"{dam}のダメージ!")
    elif com2 != e:
      eh = eh - dam
      print(f"{plname}は{maho[lev-1][com2-1]}を唱えた。")
      print(f"{dam}のダメージ!")
    else:
      hp = hp + dam
      print(f"{plname}は{maho[lev-1][com2-1]}を唱えた。")
      print(f"HPを{dam}回復")

  if eh <= 0:
    x = 0
    finish = enemy + "をたおした"
  else:
    if eh <= 80 and hoim != 2:
      he1 = pl_sm[mahoh[hoim]]
      he2 = pl_bg[mahoh[hoim]]
      he = r_mod.randint(he1, he2)
      eh = eh + he
      print(f"{enemy}は{moahoh[hoim]}を唱えた。")
      print(f"{enemy}はHPを{he}回復")
    else:
      hp = hp - enatk
      print(f"{enemy}の攻撃！")
      print(f"{enatk}のダメージを受けた")

    if hp <= 0:
      x = 0
      finish = plname + "はしんでしまった。"

print(finish)
```
